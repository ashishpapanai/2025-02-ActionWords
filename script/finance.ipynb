## 1. Temporal Evolution of Financial Control by Gender

```python
# Income control over time by gender
income_control_temporal = df_combined.groupby(['year', 'sex'])['code_income_control'].value_counts(normalize=True).reset_index()
income_control_temporal.columns = ['year', 'sex', 'income_control', 'proportion']

# Create grouped bar chart
fig1 = px.bar(income_control_temporal, 
              x='year', 
              y='proportion', 
              color='income_control',
              facet_col='sex',
              barmode='group',
              title='Financial Control Distribution by Gender Across Time',
              labels={'proportion': 'Proportion', 'year': 'Year', 'income_control': 'Income Control'},
              color_discrete_sequence=px.colors.qualitative.Set2,
              height=500)

fig1.update_layout(
    font=dict(size=12, family='Arial'),
    title_font_size=16,
    plot_bgcolor='white',
    paper_bgcolor='white',
    legend_title_text='Income Control'
)

fig1.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
fig1.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')

fig1.show()
```

```python
# Pie charts showing education distribution by gender for each year
from plotly.subplots import make_subplots

# Prepare data
edu_pie_data = df_combined.groupby(['year', 'sex', 'educ_4_label']).size().reset_index(name='count')
edu_pie_total = df_combined.groupby(['year', 'sex']).size().reset_index(name='total')
edu_pie_data = edu_pie_data.merge(edu_pie_total, on=['year', 'sex'])
edu_pie_data['proportion'] = edu_pie_data['count'] / edu_pie_data['total']

# Create subplots - 2 rows (Male/Female) x 3 columns (Years)
fig_pie = make_subplots(
    rows=2, cols=3,
    specs=[[{'type': 'pie'}, {'type': 'pie'}, {'type': 'pie'}],
           [{'type': 'pie'}, {'type': 'pie'}, {'type': 'pie'}]],
    subplot_titles=('Male - 2002', 'Male - 2012', 'Male - 2022',
                    'Female - 2002', 'Female - 2012', 'Female - 2022'),
    vertical_spacing=0.15,
    horizontal_spacing=0.05
)

# Define colors for education levels (consistent across all pie charts)
edu_colors = {
    'No degree': '#d62728',
    'Lower secondary': '#ff7f0e',
    'Upper secondary': '#2ca02c',
    'University': '#1f77b4'
}

# Add pie charts
for row_idx, gender in enumerate(['Male', 'Female'], start=1):
    for col_idx, year in enumerate([2002, 2012, 2022], start=1):
        data_subset = edu_pie_data[(edu_pie_data['sex'] == gender) & (edu_pie_data['year'] == year)]
        
        # Sort by education level for consistency
        data_subset = data_subset.sort_values('educ_4_label')
        
        colors = [edu_colors.get(edu, '#gray') for edu in data_subset['educ_4_label']]
        
        fig_pie.add_trace(
            go.Pie(
                labels=data_subset['educ_4_label'],
                values=data_subset['count'],
                marker=dict(colors=colors),
                textposition='inside',
                textinfo='percent',
                textfont=dict(size=14, color='white'),
                insidetextorientation='radial',
                hovertemplate='<b>%{label}</b><br>Percentage: %{percent}<br>Count: %{value}<extra></extra>',
                showlegend=(row_idx == 1 and col_idx == 1)  # Show legend only once
            ),
            row=row_idx, col=col_idx
        )

fig_pie.update_layout(
    title_text='Education Level Distribution by Gender and Year',
    title_font_size=20,
    font=dict(size=13, family='Arial'),
    height=1000,
    showlegend=True,
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.05,
        xanchor="center",
        x=0.5,
        font=dict(size=14)
    )
)

# Update subplot titles font size
for annotation in fig_pie['layout']['annotations']:
    annotation['font'] = dict(size=15)

fig_pie.show()
```

## 9. Sankey Diagram: Financial Control Flow by Demographics

```python
# Sankey diagram showing flow from gender -> higher earner -> income control (2022)
sankey_data = df_combined[df_combined['year'] == 2022][['sex', 'code_higher_income', 'code_income_control']].dropna()

# Create node labels
all_sources = list(sankey_data['sex'].unique()) + list(sankey_data['code_higher_income'].unique())
all_targets = list(sankey_data['code_higher_income'].unique()) + list(sankey_data['code_income_control'].unique())
all_labels = list(set(all_sources + all_targets))

# Create label to index mapping
label_to_idx = {label: idx for idx, label in enumerate(all_labels)}

# Create links
source_list = []
target_list = []
value_list = []

# Gender to Higher Earner
for idx, row in sankey_data.groupby(['sex', 'code_higher_income']).size().reset_index(name='count').iterrows():
    source_list.append(label_to_idx[row['sex']])
    target_list.append(label_to_idx[row['code_higher_income']])
    value_list.append(row['count'])

# Higher Earner to Income Control
for idx, row in sankey_data.groupby(['code_higher_income', 'code_income_control']).size().reset_index(name='count').iterrows():
    source_list.append(label_to_idx[row['code_higher_income']])
    target_list.append(label_to_idx[row['code_income_control']])
    value_list.append(row['count'])

fig9 = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color='black', width=0.5),
        label=all_labels,
        color=['#3498db', '#e74c3c', '#f39c12', '#2ecc71', '#9b59b6', 
               '#1abc9c', '#e67e22', '#34495e'][:len(all_labels)]
    ),
    link=dict(
        source=source_list,
        target=target_list,
        value=value_list
    )
)])

fig9.update_layout(
    title_text="Financial Control Flow: Gender  Higher Earner  Control Type (2022)",
    title_font_size=16,
    font=dict(size=12, family='Arial'),
    height=600
)

fig9.show()
```

## 10. Heatmap: Financial Control by Work Status and Gender

```python
# Create pivot table for heatmap
work_pivot_data = df_combined[(df_combined['year'] == 2022) & 
                                (df_combined['code_income_control'] == 'Joint')].copy()

# Calculate proportion by work status
work_status_counts = work_pivot_data.groupby(['work_status_std', 'sex']).size().reset_index(name='joint_count')
total_counts = df_combined[df_combined['year'] == 2022].groupby(['work_status_std', 'sex']).size().reset_index(name='total')
work_pivot = work_status_counts.merge(total_counts, on=['work_status_std', 'sex'])
work_pivot['proportion'] = work_pivot['joint_count'] / work_pivot['total']

# Pivot for heatmap
heatmap_data = work_pivot.pivot(index='work_status_std', columns='sex', values='proportion')

fig10 = px.imshow(heatmap_data,
                  labels=dict(x='Gender', y='Work Status', color='Proportion with<br>Joint Control'),
                  title='Joint Financial Control by Work Status and Gender (2022)',
                  color_continuous_scale='RdYlGn',
                  aspect='auto',
                  height=500)

fig10.update_layout(
    font=dict(size=12, family='Arial'),
    title_font_size=16
)

fig10.show()
```

## 2. COMPREHENSIVE FINANCIAL EQUALITY ANALYSIS: A MULTI-DIMENSIONAL STORY

This section explores financial equality through multiple lenses: household work division, life satisfaction, work patterns, social status, and more. Each visualization tells part of the story of how different factors influence financial equality in married couples.

```python
# STORY PART 1: The Household Work-Financial Control Connection
# Does sharing household work lead to sharing financial control?

# Check data availability
if 'SHARE_HH' in df_combined.columns:
    hh_finance_data = df_combined[(df_combined['SHARE_HH'].notna()) & 
                                   (df_combined['code_income_control'].notna()) &
                                   (df_combined['year'] == 2022)].copy()
    
    if len(hh_finance_data) > 0:
        # Create household work share categories
        hh_finance_data['hh_share_cat'] = pd.cut(hh_finance_data['SHARE_HH'], 
                                                  bins=[0, 25, 50, 75, 100], 
                                                  labels=['0-25%', '26-50%', '51-75%', '76-100%'],
                                                  include_lowest=True)
        
        hh_finance = hh_finance_data.groupby(['hh_share_cat', 'code_income_control', 'sex']).size().reset_index(name='count')
        hh_finance_total = hh_finance_data.groupby(['hh_share_cat', 'sex']).size().reset_index(name='total')
        hh_finance = hh_finance.merge(hh_finance_total, on=['hh_share_cat', 'sex'])
        hh_finance['proportion'] = hh_finance['count'] / hh_finance['total']
        
        # Focus on joint control
        hh_joint = hh_finance[hh_finance['code_income_control'] == 'Joint']
        
        fig_hh_finance = px.bar(hh_joint,
                                x='hh_share_cat',
                                y='proportion',
                                color='sex',
                                barmode='group',
                                title='Story 1: Household Work Share and Joint Financial Control (2022)<br><sub>Does doing more household work correlate with having joint financial control?</sub>',
                                labels={'proportion': 'Proportion with Joint Control',
                                        'hh_share_cat': 'Household Work Share (%)',
                                        'sex': 'Gender'},
                                color_discrete_map={'Male': '#3498db', 'Female': '#e74c3c'},
                                height=600)
        
        fig_hh_finance.update_layout(
            font=dict(size=12, family='Arial'),
            title_font_size=16,
            plot_bgcolor='white',
            paper_bgcolor='white'
        )
        
        fig_hh_finance.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        fig_hh_finance.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        
        fig_hh_finance.show()
    else:
        print("No data available for household work share analysis")
else:
    print("SHARE_HH column not available")
```

```python
# STORY PART 2: Life Satisfaction and Financial Equality
# Are couples with joint financial control happier?

if 'LIFE_HAP' in df_combined.columns:
    life_finance_data = df_combined[(df_combined['LIFE_HAP'].notna()) & 
                                     (df_combined['code_income_control'].notna()) &
                                     (df_combined['year'] == 2022)].copy()
    
    if len(life_finance_data) > 0:
        # Calculate average life satisfaction by financial control type
        life_by_control = life_finance_data.groupby(['code_income_control', 'sex'])['LIFE_HAP'].mean().reset_index()
        life_by_control.columns = ['code_income_control', 'sex', 'avg_life_satisfaction']
        
        fig_life_finance = px.bar(life_by_control,
                                  x='code_income_control',
                                  y='avg_life_satisfaction',
                                  color='sex',
                                  barmode='group',
                                  title='Story 2: Life Satisfaction by Financial Control Type and Gender (2022)<br><sub>Does financial equality contribute to happiness?</sub>',
                                  labels={'avg_life_satisfaction': 'Average Life Satisfaction',
                                          'code_income_control': 'Financial Control Type',
                                          'sex': 'Gender'},
                                  color_discrete_map={'Male': '#3498db', 'Female': '#e74c3c'},
                                  height=600)
        
        fig_life_finance.update_layout(
            font=dict(size=11, family='Arial'),
            title_font_size=16,
            plot_bgcolor='white',
            paper_bgcolor='white',
            xaxis_tickangle=-45
        )
        
        fig_life_finance.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        fig_life_finance.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        
        fig_life_finance.show()
    else:
        print("No data available for life satisfaction analysis")
else:
    print("LIFE_HAP column not available")
```

```python
# STORY PART 3: Work Hours and Financial Control
# Do couples where both work similar hours have more equal financial control?

if 'wrk_hrs' in df_combined.columns:
    work_finance_data = df_combined[(df_combined['wrk_hrs'].notna()) & 
                                     (df_combined['code_income_control'].notna()) &
                                     (df_combined['code_higher_income'].notna()) &
                                     (df_combined['year'] == 2022)].copy()
    
    if len(work_finance_data) > 0:
        # Focus on equal earners vs. unequal earners
        equal_earners = work_finance_data[work_finance_data['code_higher_income'] == 'Both earn about the same']
        unequal_earners = work_finance_data[work_finance_data['code_higher_income'] != 'Both earn about the same']
        
        equal_joint = equal_earners[equal_earners['code_income_control'] == 'Joint'].shape[0] / len(equal_earners) if len(equal_earners) > 0 else 0
        unequal_joint = unequal_earners[unequal_earners['code_income_control'] == 'Joint'].shape[0] / len(unequal_earners) if len(unequal_earners) > 0 else 0
        
        comparison_data = pd.DataFrame({
            'Earning Status': ['Equal Earners', 'Unequal Earners'],
            'Proportion with Joint Control': [equal_joint, unequal_joint]
        })
        
        fig_work_finance = px.bar(comparison_data,
                                  x='Earning Status',
                                  y='Proportion with Joint Control',
                                  title='Story 3: Equal Earnings and Joint Financial Control (2022)<br><sub>Do couples who earn similar amounts share financial control more?</sub>',
                                  color='Earning Status',
                                  color_discrete_map={'Equal Earners': '#2ecc71', 'Unequal Earners': '#e74c3c'},
                                  height=600)
        
        fig_work_finance.update_layout(
            font=dict(size=12, family='Arial'),
            title_font_size=16,
            plot_bgcolor='white',
            paper_bgcolor='white',
            showlegend=False
        )
        
        fig_work_finance.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        fig_work_finance.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        
        fig_work_finance.show()
    else:
        print("No data available for work hours analysis")
else:
    print("wrk_hrs column not available")
```

```python
# STORY PART 4: Household Work Division Details
# Which household tasks are most associated with financial control?

hh_tasks = ['DIV_HH_COOK', 'DIV_HH_CLEAN', 'DIV_HH_GROC', 'DIV_HH_CARE', 'DIV_HH_LAUND']
available_tasks = [task for task in hh_tasks if task in df_combined.columns]

if len(available_tasks) > 0:
    task_data = df_combined[(df_combined['code_income_control'].notna()) & 
                            (df_combined['year'] == 2022)].copy()
    
    if len(task_data) > 0:
        task_results = []
        for task in available_tasks:
            task_subset = task_data[task_data[task].notna()]
            if len(task_subset) > 0:
                # Calculate proportion of joint control for different task divisions
                task_joint = task_subset[task_subset['code_income_control'] == 'Joint'].shape[0] / len(task_subset)
                task_results.append({
                    'Task': task.replace('DIV_HH_', '').replace('_', ' ').title(),
                    'Proportion Joint Control': task_joint,
                    'Sample Size': len(task_subset)
                })
        
        if len(task_results) > 0:
            task_df = pd.DataFrame(task_results)
            
            fig_tasks = px.bar(task_df,
                              x='Task',
                              y='Proportion Joint Control',
                              title='Story 4: Household Task Division and Financial Control (2022)<br><sub>Which household responsibilities correlate with joint financial control?</sub>',
                              color='Proportion Joint Control',
                              color_continuous_scale='RdYlGn',
                              height=600)
            
            fig_tasks.update_layout(
                font=dict(size=12, family='Arial'),
                title_font_size=16,
                plot_bgcolor='white',
                paper_bgcolor='white',
                xaxis_tickangle=-45
            )
            
            fig_tasks.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
            fig_tasks.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
            
            fig_tasks.show()
        else:
            print("No valid task data available")
    else:
        print("No data available for task analysis")
else:
    print("No household task columns available")
```

```python
# STORY PART 5: Social Status (TOPBOT) and Financial Equality
# Does perceived social status influence financial control patterns?

if 'TOPBOT' in df_combined.columns:
    # Create social status categories
    topbot_data = df_combined[(df_combined['TOPBOT'].notna()) & 
                               (df_combined['code_income_control'].notna()) &
                               (df_combined['year'] == 2022)].copy()
    
    if len(topbot_data) > 0:
        topbot_data['topbot_category'] = pd.cut(topbot_data['TOPBOT'], 
                                                 bins=[0, 3, 5, 7, 10], 
                                                 labels=['Low (1-3)', 'Medium-Low (4-5)', 'Medium-High (6-7)', 'High (8-10)'],
                                                 include_lowest=True)
        
        topbot_finance = topbot_data.groupby(['topbot_category', 'code_income_control', 'sex']).size().reset_index(name='count')
        topbot_total = topbot_data.groupby(['topbot_category', 'sex']).size().reset_index(name='total')
        topbot_finance = topbot_finance.merge(topbot_total, on=['topbot_category', 'sex'])
        topbot_finance['proportion'] = topbot_finance['count'] / topbot_finance['total']
        
        # Focus on joint control
        topbot_joint = topbot_finance[topbot_finance['code_income_control'] == 'Joint']
        
        fig_topbot = px.bar(topbot_joint,
                           x='topbot_category',
                           y='proportion',
                           color='sex',
                           barmode='group',
                           title='Story 5: Social Status and Joint Financial Control (2022)<br><sub>TOPBOT: Self-reported social status (1=Lowest, 10=Highest)</sub>',
                           labels={'proportion': 'Proportion with Joint Control',
                                   'topbot_category': 'Social Status Category',
                                   'sex': 'Gender'},
                           color_discrete_map={'Male': '#3498db', 'Female': '#e74c3c'},
                           height=600)
        
        fig_topbot.update_layout(
            font=dict(size=12, family='Arial'),
            title_font_size=16,
            plot_bgcolor='white',
            paper_bgcolor='white'
        )
        
        fig_topbot.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray', tickangle=45)
        fig_topbot.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        
        fig_topbot.show()
    else:
        print("No data available for TOPBOT analysis")
else:
    print("TOPBOT column not available")
```

```python
# STORY PART 6: Tiredness and Financial Control
# Does work-life balance affect financial control decisions?

if 'WORK_TIRED' in df_combined.columns or 'HH_TIRED' in df_combined.columns:
    tired_vars = []
    if 'WORK_TIRED' in df_combined.columns:
        tired_vars.append('WORK_TIRED')
    if 'HH_TIRED' in df_combined.columns:
        tired_vars.append('HH_TIRED')
    
    tired_data = df_combined[(df_combined['code_income_control'].notna()) & 
                              (df_combined['year'] == 2022)].copy()
    
    if len(tired_data) > 0 and len(tired_vars) > 0:
        tired_results = []
        for var in tired_vars:
            var_subset = tired_data[tired_data[var].notna()]
            if len(var_subset) > 0:
                # Calculate joint control proportion by tiredness level
                tired_by_control = var_subset.groupby([var, 'code_income_control']).size().reset_index(name='count')
                tired_total = var_subset.groupby(var).size().reset_index(name='total')
                tired_merged = tired_by_control.merge(tired_total, on=var)
                tired_merged['proportion'] = tired_merged['count'] / tired_merged['total']
                
                tired_joint = tired_merged[tired_merged['code_income_control'] == 'Joint']
                if len(tired_joint) > 0:
                    tired_results.append({
                        'Type': var.replace('_', ' ').title(),
                        'Data': tired_joint
                    })
        
        if len(tired_results) > 0:
            # Use the first available variable
            plot_data = tired_results[0]['Data']
            var_name = tired_results[0]['Type']
            
            fig_tired = px.bar(plot_data,
                              x=var_name.replace(' ', '_').lower(),
                              y='proportion',
                              title=f'Story 6: Tiredness and Joint Financial Control (2022)<br><sub>Does work-life balance affect financial equality?</sub>',
                              labels={'proportion': 'Proportion with Joint Control',
                                      var_name.replace(' ', '_').lower(): var_name},
                              color='proportion',
                              color_continuous_scale='RdYlGn',
                              height=600)
            
            fig_tired.update_layout(
                font=dict(size=12, family='Arial'),
                title_font_size=16,
                plot_bgcolor='white',
                paper_bgcolor='white'
            )
            
            fig_tired.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
            fig_tired.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
            
            fig_tired.show()
        else:
            print("No valid tiredness data available")
    else:
        print("No data available for tiredness analysis")
else:
    print("Tiredness columns not available")
```

```python
# STORY PART 7: Education and Financial Control Over Time
# How has the relationship between education and financial equality evolved?

if 'educ_4_label' in df_combined.columns:
    edu_temporal_data = df_combined[(df_combined['educ_4_label'].notna()) & 
                                     (df_combined['code_income_control'].notna())].copy()
    
    if len(edu_temporal_data) > 0:
        edu_temporal = edu_temporal_data[edu_temporal_data['code_income_control'] == 'Joint'].groupby(['educ_4_label', 'year', 'sex']).size().reset_index(name='joint_count')
        edu_temporal_total = edu_temporal_data.groupby(['educ_4_label', 'year', 'sex']).size().reset_index(name='total')
        edu_temporal = edu_temporal.merge(edu_temporal_total, on=['educ_4_label', 'year', 'sex'], how='right')
        edu_temporal['proportion'] = edu_temporal['joint_count'].fillna(0) / edu_temporal['total']
        
        edu_temporal['year'] = edu_temporal['year'].astype(str)
        
        fig_edu_temporal = px.line(edu_temporal,
                                   x='year',
                                   y='proportion',
                                   color='educ_4_label',
                                   line_dash='sex',
                                   markers=True,
                                   title='Story 7: Education and Joint Financial Control Over Time<br><sub>How has education influenced financial equality across decades?</sub>',
                                   labels={'proportion': 'Proportion with Joint Control',
                                           'year': 'Year',
                                           'educ_4_label': 'Education Level',
                                           'sex': 'Gender'},
                                   height=700)
        
        fig_edu_temporal.update_layout(
            font=dict(size=12, family='Arial'),
            title_font_size=16,
            plot_bgcolor='white',
            paper_bgcolor='white',
            legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
        )
        
        fig_edu_temporal.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        fig_edu_temporal.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        
        fig_edu_temporal.show()
    else:
        print("No data available for education temporal analysis")
else:
    print("educ_4_label column not available")
```

```python
# STORY PART 8: Family Structure and Financial Control
# How do children and household composition affect financial equality?

family_vars = ['HHCHILDR', 'HHADULT', 'HOMPOP']
available_family = [var for var in family_vars if var in df_combined.columns]

if len(available_family) > 0:
    family_data = df_combined[(df_combined['code_income_control'].notna()) & 
                               (df_combined['year'] == 2022)].copy()
    
    if len(family_data) > 0:
        # Use HHCHILDR if available, otherwise HOMPOP
        var_to_use = 'HHCHILDR' if 'HHCHILDR' in available_family else available_family[0]
        family_subset = family_data[family_data[var_to_use].notna()].copy()
        
        if len(family_subset) > 0:
            # Create categories
            if var_to_use == 'HHCHILDR':
                family_subset['children_cat'] = family_subset[var_to_use].apply(
                    lambda x: 'No Children' if x == 0 else ('1-2 Children' if x <= 2 else '3+ Children')
                )
                cat_col = 'children_cat'
                x_label = 'Number of Children'
            else:
                family_subset['household_cat'] = pd.cut(family_subset[var_to_use], 
                                                        bins=[0, 2, 4, 10], 
                                                        labels=['Small (1-2)', 'Medium (3-4)', 'Large (5+)'],
                                                        include_lowest=True)
                cat_col = 'household_cat'
                x_label = 'Household Size'
            
            family_finance = family_subset[family_subset['code_income_control'] == 'Joint'].groupby([cat_col, 'sex']).size().reset_index(name='joint_count')
            family_total = family_subset.groupby([cat_col, 'sex']).size().reset_index(name='total')
            family_merged = family_finance.merge(family_total, on=[cat_col, 'sex'], how='right')
            family_merged['proportion'] = family_merged['joint_count'].fillna(0) / family_merged['total']
            
            fig_family = px.bar(family_merged,
                               x=cat_col,
                               y='proportion',
                               color='sex',
                               barmode='group',
                               title=f'Story 8: Family Structure and Joint Financial Control (2022)<br><sub>How does household composition affect financial equality?</sub>',
                               labels={'proportion': 'Proportion with Joint Control',
                                       cat_col: x_label,
                                       'sex': 'Gender'},
                               color_discrete_map={'Male': '#3498db', 'Female': '#e74c3c'},
                               height=600)
            
            fig_family.update_layout(
                font=dict(size=12, family='Arial'),
                title_font_size=16,
                plot_bgcolor='white',
                paper_bgcolor='white'
            )
            
            fig_family.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
            fig_family.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
            
            fig_family.show()
        else:
            print("No valid family data available")
    else:
        print("No data available for family analysis")
else:
    print("Family structure columns not available")
```

```python
# STORY PART 9: Comprehensive Multi-Factor Analysis
# A heatmap showing how multiple factors interact with financial control

# Prepare data with multiple factors
multi_factor_data = df_combined[(df_combined['code_income_control'].notna()) & 
                                  (df_combined['year'] == 2022)].copy()

if len(multi_factor_data) > 0:
    # Create factor categories
    factors = {}
    
    if 'educ_4_label' in multi_factor_data.columns:
        factors['Education'] = multi_factor_data['educ_4_label']
    
    if 'code_higher_income' in multi_factor_data.columns:
        multi_factor_data['earner_cat'] = multi_factor_data['code_higher_income'].apply(
            lambda x: 'Equal' if 'same' in str(x).lower() else 'Unequal'
        )
        factors['Earning Status'] = multi_factor_data['earner_cat']
    
    if len(factors) >= 2:
        # Create a 2D analysis
        factor1_name = list(factors.keys())[0]
        factor2_name = list(factors.keys())[1]
        
        multi_factor_data[factor1_name] = factors[factor1_name]
        multi_factor_data[factor2_name] = factors[factor2_name]
        
        heatmap_data = multi_factor_data.groupby([factor1_name, factor2_name, 'code_income_control']).size().reset_index(name='count')
        heatmap_total = multi_factor_data.groupby([factor1_name, factor2_name]).size().reset_index(name='total')
        heatmap_merged = heatmap_data.merge(heatmap_total, on=[factor1_name, factor2_name])
        heatmap_merged['proportion'] = heatmap_merged['count'] / heatmap_merged['total']
        
        heatmap_joint = heatmap_merged[heatmap_merged['code_income_control'] == 'Joint']
        
        if len(heatmap_joint) > 0:
            heatmap_pivot = heatmap_joint.pivot(index=factor1_name, columns=factor2_name, values='proportion')
            
            fig_multi = px.imshow(heatmap_pivot,
                                 labels=dict(x=factor2_name, y=factor1_name, color='Proportion with<br>Joint Control'),
                                 title='Story 9: Multi-Factor Analysis of Financial Equality (2022)<br><sub>How do education and earning status interact?</sub>',
                                 color_continuous_scale='RdYlGn',
                                 aspect='auto',
                                 height=600,
                                 text_auto='.2f')
            
            fig_multi.update_layout(
                font=dict(size=12, family='Arial'),
                title_font_size=16
            )
            
            fig_multi.show()
        else:
            print("No joint control data for heatmap")
    else:
        print("Not enough factors available for multi-factor analysis")
else:
    print("No data available for multi-factor analysis")
```

```python
# STORY PART 10: The Complete Picture - Temporal Evolution Across All Factors
# How has financial equality changed over time, and what factors drive it?

temporal_story = df_combined[df_combined['code_income_control'].notna()].copy()

if len(temporal_story) > 0:
    # Calculate joint control proportion over time by key factors
    temporal_factors = {}
    
    if 'educ_4_label' in temporal_story.columns:
        edu_temporal = temporal_story[temporal_story['code_income_control'] == 'Joint'].groupby(['educ_4_label', 'year']).size().reset_index(name='joint_count')
        edu_temporal_total = temporal_story.groupby(['educ_4_label', 'year']).size().reset_index(name='total')
        edu_temporal = edu_temporal.merge(edu_temporal_total, on=['educ_4_label', 'year'], how='right')
        edu_temporal['proportion'] = edu_temporal['joint_count'].fillna(0) / edu_temporal['total']
        edu_temporal['year'] = edu_temporal['year'].astype(str)
        
        fig_temporal = px.line(edu_temporal,
                               x='year',
                               y='proportion',
                               color='educ_4_label',
                               markers=True,
                               title='Story 10: The Evolution of Financial Equality (2002-2022)<br><sub>How has joint financial control changed across education levels over two decades?</sub>',
                               labels={'proportion': 'Proportion with Joint Control',
                                       'year': 'Year',
                                       'educ_4_label': 'Education Level'},
                               height=700)
        
        fig_temporal.update_layout(
            font=dict(size=12, family='Arial'),
            title_font_size=16,
            plot_bgcolor='white',
            paper_bgcolor='white',
            legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
        )
        
        fig_temporal.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        fig_temporal.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        
        fig_temporal.show()
    else:
        # Fallback: just by gender over time
        gender_temporal = temporal_story[temporal_story['code_income_control'] == 'Joint'].groupby(['sex', 'year']).size().reset_index(name='joint_count')
        gender_temporal_total = temporal_story.groupby(['sex', 'year']).size().reset_index(name='total')
        gender_temporal = gender_temporal.merge(gender_temporal_total, on=['sex', 'year'], how='right')
        gender_temporal['proportion'] = gender_temporal['joint_count'].fillna(0) / gender_temporal['total']
        gender_temporal['year'] = gender_temporal['year'].astype(str)
        
        fig_temporal = px.line(gender_temporal,
                               x='year',
                               y='proportion',
                               color='sex',
                               markers=True,
                               title='Story 10: The Evolution of Financial Equality (2002-2022)<br><sub>How has joint financial control changed over two decades?</sub>',
                               labels={'proportion': 'Proportion with Joint Control',
                                       'year': 'Year',
                                       'sex': 'Gender'},
                               color_discrete_map={'Male': '#3498db', 'Female': '#e74c3c'},
                               height=600)
        
        fig_temporal.update_layout(
            font=dict(size=12, family='Arial'),
            title_font_size=16,
            plot_bgcolor='white',
            paper_bgcolor='white'
        )
        
        fig_temporal.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        fig_temporal.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
        
        fig_temporal.show()
else:
    print("No data available for temporal analysis")
```
